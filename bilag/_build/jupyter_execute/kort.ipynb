```python
df = pd.DataFrame(data)
death_data = pd.read_excel('Book1.xlsx')
line_points = df[['Latitude', 'Longitude']].values.tolist()

min_lat = df['Latitude'].min()
max_lat = df['Latitude'].max()
min_lon = df['Longitude'].min()
max_lon = df['Longitude'].max()

# Design
m = folium.Map(location=[(min_lat + max_lat) / 3, (min_lon + max_lon) / 3], zoom_start=3, tiles=None)
folium.TileLayer(
    tiles='https://services.arcgisonline.com/arcgis/rest/services/World_Shaded_Relief/MapServer/tile/{z}/{y}/{x}',
    attr='Esri World Shaded Relief',
    name='Vintage',
    control=False
).add_to(m)

# Linje
folium.PolyLine(line_points, color="grey", weight=3, opacity=0.7).add_to(m)
m.fit_bounds([[min_lat, min_lon], [max_lat, max_lon]])

# Placering af "plots"
journey_duration = 175
for index, row in death_data.iterrows():
    days = row['days_taken']
    # Ensure the days are within the journey duration
    if days <= journey_duration:
        position_index = int((days / journey_duration) * (len(line_points) - 1))
        death_location = line_points[position_index]
        folium.CircleMarker(
            location=death_location,
            radius=5,
            color='red',
            fill=True,
            fill_color='red'
        ).add_to(m)
m
```
